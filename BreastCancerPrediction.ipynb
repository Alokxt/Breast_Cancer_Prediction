Let's do a kruskal-wallis test to test the significance of each column

```python
from spicy import stats
```

```python
def kruskal_wallis(df):
    column_to_test = list(df.select_dtypes('float64').columns)
    useless_cols = []
    for i in column_to_test:
        mels = df[i][df['diagnosis'] == 'M']
        ben = df[i][df['diagnosis'] == 'B']

        stat_val , p_value = stats.kruskal(mels,ben)
        if p_value >=0.05:
            print(p_value,i)
            useless_cols.append(i)
    return useless_cols
```

```python
cov_mat = np.cov(X_centred.T)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_mat)
```

```python
eigen_values.shape
```

```python
eigen_vectors.shape
```

```python

total = np.sum(eigen_values)
print(total)
```

```python
eigen_values_ratio = (eigen_values / total)*100
print(eigen_values_ratio)
```

```python
explained_var =[]
c = 0
for i , var in enumerate(eigen_values_ratio):
    c += round(var,3)
    if c<92:
          explained_var.append(var)
          print(f'At {i+1} PC, Explained Variance is {round(c,2)}%')
    else:
        explained_var.append(var)
        print(f'At {i+1} PC, Explained Variance crosses threshold: {round(c,2)}%')
        break
    
    
```

So we have to take 8 principle components

```python
explained_variance_ratio = eigen_values / np.sum(eigen_values) * 100
cumulative_variance = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Explained Variance vs Number of Components')
plt.grid(True)
plt.show()
```
